# Importación de Datos en Python

El primer paso para importar datos de Google Forms, es importar la base de datos que se crea automáticamente en Google Sheets. 

Desde Python hay una manera bastante rápida y simple para poder hacer esto:

```python
import pandas as pd

sheet_url = 'https://docs.google.com/spreadsheets/d/1qNgkIIXrpT_TW8LZ574kq1ogn1ns4kgYNXNjILiMegw/edit#gid=0'
url_1 = sheet_url.replace('/edit#gid=', '/export?format=csv&gid=')
df = pd.read_csv(url_1, decimal=",")
df
```

¡Listo! Ya tenemos nuestra base de datos. Sin embargo, van a notar que todavía no está lista para poder trabajarse. Es necesario que hagamos los siguientes cambios:

1. Proponer nombres de variables mas simples
2. Cambiar de letras a números las escalas likerts de los ítems
3. Invertir ítems y calcular los puntajes de las variables

## Cambiar nombres de variables

Para cambiar los nombres de las variables, primero es importante que podamos ver el nombre que tienen en este momento. Para eso, podemos hacer uso del comando "columns" del Data Frame:

```python
df.columns
```

Con esta información, podemos usar el comando "columns" para introducir los nuevos nombres de nuestras variables:

```python
df.columns = ["id", "clase", "grado", "rend", "mot",
            "c1_1a", "c1_2a", "c1_3a", "c1_1b", "c1_2b", "c1_3b",
            "c2_1", "c2_2", "c2_3", "c2_4", "c2_5"]
df
```

¡Listo! con los nombres abreviados, ahora podemos pasar a cambiar de letras a números las escalas likerts de nuestros ítems.

## De letras a números

A diferencia de otros softwares, Python es lo suficiente inteligente para poder trabajar con las variables que usan letras de manera eficiente. No obstante, si tenemos que generar un puntaje o deseamos tratar estas variables como continuas, necesitamos transformarlas a números.

En este caso, necesitamos hacer esta transformación con los ítems de los cuestionarios para poder generar los puntajes de nuestras escalas:

```python
claves_c1 = {"En desacuerdo": 1.0,
          "Neutral": 2.0,
          "De acuerdo": 3.0,
          "Muy de acuerdo": 4.0}

claves_c2 = {"(1) Siempre": 1.0,
             "(2) Casi siempre": 2.0,
             "(3) Algunas veces": 3.0,
             "(4) Casi nunca": 4.0,
             "(5) Nunca": 5.0}

df.replace(claves_c1, inplace=True)
df.replace(claves_c2, inplace=True)

df
```

## Invertir items

Muchas veces los instrumentos vienen con algunos ítems inversos. En esos casos, necesitamos invertir su dirección para poder trabajar con ellos. 

En Python podemos hacer una simple resta para poder hacer esta inversión. Para esto restamos los items del valor total posible de la escala likert + 1.

```python
df["c1_1ai"] = 5 - df["c1_1a"]
df["c1_2i"] = 6 - df["c2_2"]
df
```

Con esto, le hemos pedido a Python que invierta los ítems c1_a y c2_2 y guarde este resultado en dos variables nuevas que terminan con una "i".

Tengan en cuenta que esta inversión funcionó porque el menor valor de ambos instrumentos es "1". Si este fuese 0, tendríamos que restar el número exacto de puntos de la escala likert (p.e.: `df["c1_2i"] = 6 - df["c2_2"]`).

¡Listo! ahora pueden pasar a crear sus variables e iniciar sus análisis.
